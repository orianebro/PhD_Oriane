```python
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_simpson, simpson
from openalea.archicrop.cereals_leaf import parametric_leaf
```

```python
x,y,s,r = parametric_leaf(
        nb_segment=100,
        insertion_angle=30,
        scurv=0.7,
        curvature=100,
        # klig=0.6, swmax=0.55, f1=0.64, f2=0.92 # wheat
        klig=0.6, swmax=0.5, f1=0.4, f2=0.6
)

r
```

```python
r = r[::-1]
r
```

```python
wl = 0.12
r *= wl
r
```

```python
S = cumulative_simpson(x=s, y=r, initial=0)
S
```

```python
S_norm = S/S[-1]
S_norm
```

```python
S_max = 10
S_scaled = S_norm * S_max
S_scaled
```

```python
plt.figure(figsize=(10, 1))
S_scaled_uncum = [S_scaled[0]] + [S_scaled[i+1]-S_scaled[i] for i in range(len(S_scaled[1:]))]
print(len(S_scaled_uncum))
plt.plot(range(len(S_scaled_uncum)),S_scaled_uncum)
plt.show()
```

```python
plt.figure(figsize=(10, 1))
plt.plot(s,r)
plt.show()
```

```python
from scipy.interpolate import splrep, splev

t,c,k = splrep(x=S_norm, y=s, k=3, task=0)

y = splev(x=S_norm, tck=(t,c,k))

plt.plot(S_norm, y)
plt.plot(s, S_norm)
plt.show()
```

```python
y = splev(x=S_scaled, tck=(t,c,k))
yns = splev(x=S_norm, tck=(t,c,k))
yns *= S_max

plt.plot(S_scaled, yns)
# plt.plot(s*10, S_scaled)
plt.show()
```
